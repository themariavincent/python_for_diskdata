```python
import glob
import os
from astropy.io import fits
from astropy.table import QTable, Table
import astropy.units as u
```

```python
data_dir = '/Volumes/Maria 2020/SCExAO Data/Apr 1/'
sub_dir1 = 'charis_data/'
sub_dir2 = 'vampires_data/'
charis_files = [os.path.basename(x) for x in glob.glob(data_dir+sub_dir1+'*.fits')]
```

```python
file_name,object_name,hwp_angle=[],[],[]
for i in range(0,len(charis_files)):
    if i != 222:
        data = fits.open(data_dir+sub_dir1+charis_files[i])
        header = data[-1].header
        file_name.append(charis_files[i])
        object_name.append(header['OBJECT'])
    #     UT_date.append(header['DATE-OBS'])
        hwp_angle.append(header['P_RTAGL1'])
        # exp_time.append(header['EXPTIME'])
    else:
        continue

table = QTable([file_name, object_name,hwp_angle], names=('FILE', 'OBJECT','HWP_ANG'))
table.show_in_notebook(display_length=10)
```

```python
data = Table()
data['FILE'] = file_name
data['OBJECT'] = object_name
data['HWP_ANG'] = hwp_angle
# data['UT TIME'] = UT_time
data.write('hwp_info.csv', format='ascii', overwrite=True)  
```
